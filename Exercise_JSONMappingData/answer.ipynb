```python
#Create a variable to hold the data that only has White and Red wine. 
red_white = df[(df['variety'] == 'White Wine') | (df['variety'] == 'Red Wine')] 
red_white
```

```python
#Create an agg, grouped by variety to see the mean and median for red and white wines. 
baseline_ratings = red_white.groupby(['variety']).agg({'rating':['mean','median']})
print(baseline_ratings)
```

```python
# Based on the baseline ratings found above, create a new varible to hold the white and red wine with the ratings. 
baselined_wine_score = red_white[(red_white['rating'] >= 90) & (red_white['rating'] < 92)]
baselined_wine_score.head()
```

```python
#Group by region. 
region_baselined_wine_score = baselined_wine_score.groupby(['region'])
region_baselined_wine_score.head()
```

```python
# Take the grouped by region, filtered by red and white wine, and the baselined ratings and take the aggregrated stats below:
aggregated_stats = region_baselined_wine_score.agg({'rating':['mean', 'median', 'min', 'max']})
aggregated_stats
aggregated_stats_json = aggregated_stats.to_json()



```

```python
# Import json lib.
import json

data = json.loads(aggregated_stats_json)
formatted_data = json.dumps(data, sort_keys=True, indent=4)

with open('sample_data/test.json', 'w') as f:
    json.dump(formatted_data, f)
```
